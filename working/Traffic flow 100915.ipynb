```python
def traffic_A(t):
    v_max=80
    L=11.0
    rho_max=250
    nx=51
    dx=L/nx
    dt=0.001
    x=numpy.linspace(0,L,nx)
    v=numpy.empty(nx)
    F=numpy.empty(nx)
    rho0=numpy.ones(nx)*10
    rho0[10:20]=50
    rho=rho0
    v[0]=v_max*(1-rho0[0]/rho_max)
    F[0]=v[0]*rho0[0]
    nt=int(t/60/dt)
    for n in range(1,nt):
        rhon=rho.copy()
        for i in range(1,nx):
            v[i]=v_max*(1-rho[i]/rho_max)
            F[i]=v[i]*rho[i]
            rho[i]=rhon[i]-dt/dx*(F[i]-F[i-1])
            
    pyplot.figure(figsize=(15,5))
    pyplot.subplot(121)
    pyplot.xlabel('$x$')
    pyplot.ylabel('density')
    pyplot.plot(x,rho,'k-',lw=2)
    pyplot.title('Traffic density at t={} minutes in cars/km'.format(t),fontsize=18)
    pyplot.xlim(0,11)
    pyplot.ylim(0,50)

    pyplot.subplot(122)
    pyplot.xlabel('$x$')
    pyplot.ylabel('$v$')
    pyplot.plot(x,v,'k-',lw=2)
    pyplot.title('Velocity at t={} minutes in km/h'.format(t),fontsize=18)
    pyplot.xlim(0,11)
    
    v_total=0
    for i in range(nx):
        v_total=v_total+v[i]
    v_ave=v_total/nx
    print('The average velocity at t={} minutes is {:.2f} m/s.'.format(t,v_ave/3.6))
    
    v_min=v[0]
    for i in range(nx):
        if v[i]<v_min:
            v_min=v[i]
    print('The minimum velocity at t={} minutes is {:.2f} m/s.'.format(t,v_min/3.6))
```

```python
traffic_A(3)
```

```python
traffic_A(6)
```
