```python
import numpy
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams
rcParams['font.family']='serif'
rcParams['font.size']=16
```

```python
L=1
nt=100
nx=51
alpha=1.22e-3
dx=L/(nx-1)
Ti=numpy.zeros(nx)
Ti[0]=100
```

```python
def ftcs(T,nt,dt,dx,alpha):
    for n in range(nt):
        Tn=T.copy()
        T[1:-1]=Tn[1:-1]+alpha*dt/dx**2*(Tn[2:]-2*Tn[1:-1]+Tn[:-2])
    return T
```

```python
sigma=1/2.0
dt=sigma*dx*dx/alpha
T=ftcs(Ti.copy(),nt,dt,dx,alpha)
```

```python
pyplot.plot(numpy.linspace(0,1,nx),T,color='#003366',ls='-',lw=3)
pyplot.ylim=(0,100)
pyplot.xlabel('Length of Rod')
pyplot.ylabel('Temperature');
```

```python
nt=1000
T=ftcs(Ti.copy(),nt,dt,dx,alpha)
pyplot.plot(numpy.linspace(0,1,nx),T,color='#003366',ls='-',lw=3)
pyplot.ylim=(0,100)
pyplot.xlabel('Length of Rod')
pyplot.ylabel('Temperature');
```

```python
def ftcs_mixed(T,nt,dt,dx,alpha):
    for n in range(nt):
        Tn=T.copy()
        T[1:-1]=Tn[1:-1]+alpha*dt/dx**2*(Tn[2:]-2*Tn[1:-1]+Tn[:-2])
        T[-1]=T[-2]
    return T
```

```python
nt=1000
T=ftcs(Ti.copy(),nt,dt,dx,alpha)
pyplot.plot(numpy.linspace(0,1,nx),T,color='#003366',ls='-',lw=3)
pyplot.ylim=(0,100)
pyplot.title('Rod with LHS Dirichlet B.C. and RHS Neumann B.C. \n')
pyplot.xlabel('Length of Rod')
pyplot.ylabel('Temperature');
```
